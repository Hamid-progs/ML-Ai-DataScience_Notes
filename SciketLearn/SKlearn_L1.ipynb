# 1.Getting the data ready 

```python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("cleanHeartData.csv")
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis= 1 , inplace=True)
```

```python
df.head()
```

```python
df.shape
```

### predict from features/base data : 
age , sex  , resting_blood_pressure , cholestoral , fasting_blood_sugar , Max_heart_rate,exercise_induced_angina, oldpeak
### what is the target?
target(y) depends on base data (x)

```python
x = df.drop('target',axis=1)   # base data / features      / x is a independent variable
```

```python
y = df['target']                # target data set    / y is a dependent variable 
```

```python
x
```

```python
y
```

# 2.choosing machine learning model 

```python
 
clf = RandomForestClassifier()
```

```python
clf.get_params()
```

# 3.fit model

```python

x_train , x_test , y_train , y_test  = train_test_split(x,y, test_size=0.3)
```

```python
clf.fit(x_train,y_train)
```

# 4.evaluate model

```python
predicted_y = clf.predict(x_test)   # test data
```

```python
predicted_y       # predicted value
```

```python
clf.score(x_train,y_train)  # prediction is 100 % correct on train data
```

```python
clf.score(x_test,y_test)  # it predicted  78% correct on test data
```

# 5.improve model

we will use only 'n_estimators': 100, parameter

```python
for i in range(10,200, 10):
    print(f"Running model with {i}_estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Accuracy is : {clf.score(x_test,y_test)}")
```

Running model with 60_estimators
Accuracy is : 0.7912087912087912

```python
clf = RandomForestClassifier(n_estimators=90).fit(x_train,y_train)
print(f'{clf.score(x_test,y_test)}')
```

# 6.saving model

```python
import pickle
pickle.dump(clf, open("Heart disease Predictator.pkl",'wb'))
```

```python
load_model = pickle.load(open("Heart disease Predictator.pkl",'rb'))
load_model.score(x_test,y_test)
```
