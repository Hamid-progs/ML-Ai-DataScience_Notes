```python
df=pd.read_csv('HeartDiseaseTrain-Test.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["chest_pain_type"].unique()
```

```python
#encoding 'Typical angina', 'Atypical angina', 'Non-anginal pain','Asymptomatic' 
pd.get_dummies(df["chest_pain_type"])
df["chest_pain_type"] = df["chest_pain_type"].map({'Typical angina': 1 ,
                                                   'Atypical angina': 2,
                                                   'Non-anginal pain' : 3,
                                                   'Asymptomatic' : 4
                                                  })
```

```python
#encoding male and female
pd.get_dummies(df['sex'])
df['sex'] = df['sex'].map({'Male': 1 , 'Female':0})
```

```python
# encoding fasting blood suger
pd.get_dummies(df['fasting_blood_sugar'])
df['fasting_blood_sugar'] = df['fasting_blood_sugar'].map({'Lower than 120 mg/ml':0 ,
                                                          'Greater than 120 mg/ml':1})
```

```python
#encoding rest_ecg
pd.get_dummies(df['rest_ecg'])
df['rest_ecg']=df['rest_ecg'].map({'Normal':0 ,'Left ventricular hypertrophy':1 ,
                                  'ST-T wave abnormality':2})
```

```python
#encoding exercise_induced_angina
pd.get_dummies(df['exercise_induced_angina'])
df['exercise_induced_angina'] = df['exercise_induced_angina'].map({'Yes': 1 , 'No': 0})
```

```python
# encoding slope
pd.get_dummies(df['slope'])
df['slope'] = df['slope'].map({'Flat':0 ,'Upsloping':1,'Downsloping':2 })
```

```python
# encoding vessels_colored_by_flourosopy 
pd.get_dummies(df['vessels_colored_by_flourosopy'])
df['vessels_colored_by_flourosopy'] = df['vessels_colored_by_flourosopy'].map({'Zero':0,'Four':4,
                                                                              'One':1,'Two':2,'Three':3})
```

```python
# encoding thalassemia
pd.get_dummies(df['thalassemia'])
df['thalassemia']=df['thalassemia'].map({'No':0 , 'Normal' : 2 ,'Fixed Defect':1,'Reversable Defect':3})
```

```python
#split x and y
x = df.drop("Price", axis= 1)
y = df["Price"]
```

```python
x
```

```python
#imputation
#filling data with sk learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#fill catogorical values with misssing and numeric values with mean
#cat_features = SimpleImputer(strategy="constant", fill_value='missing')
num_features = SimpleImputer(strategy="mean")

#define columns 
num_feature = ['Battery_mAh','Storage_GB','RAM_GB','Weight_g','Rating']
imputer = ColumnTransformer([('num_features',num_features,num_feature)])
filled_x = imputer.fit_transform(x)
filled_x
```

```python
from sklearn.datasets import fetch_california_housing
c_housing = fetch_california_housing()
c_housing
```

```python
# Convert to DataFrame
df = pd.DataFrame(c_housing.data, columns=c_housing.feature_names)
# Add target column
df["target"] = c_housing['target']

# Show the first few rows
df.head()
```

```python
df.shape
```

```python
len(df)
```

### use sklearn estimator choice documentation on google

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
#to produce similar results for others
np.random.seed()
#create data 
x = df.drop('target',axis=1)
y = df['target']
# split train text 
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2)
# use instantiate , Ridge model
reg_model = Ridge()
model = reg_model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
# imported linear svc model
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

#spliting data 
x = df.drop('target',axis=1)
y = df['target']

#split train test
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)

# instinciate model
lsvc = LinearSVC()
model = lsvc.fit(x_train,y_train)

# model score
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
# the train test split is already defined above

#split data 
x = df.drop('target',axis=1)
y = df['target']

#split train test
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)

# model fitting
rfc = RandomForestClassifier()
model = rfc.fit(x_train,y_train)

# model score 
model.score(x_test,y_test)
```
