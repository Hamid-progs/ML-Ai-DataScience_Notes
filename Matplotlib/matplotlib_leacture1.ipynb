# Methods of Ploting

### M1

```python
fig = plt.figure()
sub_fig = fig.add_axes([1,1,1,1])
sub_fig.plot(x,y)
plt.show()
```

### M2

```python
#1-prepare data
x = [1,2,3,4,5,6,7]
y = [12,34,23,84,23,89,1]
#2-setup the enviornment
home = plt.figure()
room = home.add_subplot()
#3-plot the graph
room.plot(x,y)
#4-costumization
room.set(title = 'A Good plot',
         xlabel = 'x-axis',
         ylabel = 'y-axis')
#5-display plot
plt.show()
```

### M3

```python
home,room = plt.subplots()
room.plot(x,y)
plt.show()
```

# pyplot vs object oriented Method

```python
heart = pd.read_csv("HeartDiseaseTrain-Test.csv")
```

```python
heart
```

```python
heart.head()
```

```python
m = heart['age'].median()
```

```python
m
```

```python
above_55 = heart.query(f'age >= {m}')
below_55 = heart.query(f'age < {m}')
```

```python
above_55
```

```python
above_55.sort_values(by = 'age', inplace=True)
```

```python
above_55
```

```python
# pyplot method
above_55.plot(kind = 'scatter', 
             x='age',
             y= 'cholestoral', 
             c= 'target')
plt.show()
```

```python
# object oriented approach
# professional way and recommended
fig , subfig = plt.subplots(figsize=(6,5))
above_55.plot(kind='scatter',
             x='age', 
             y = 'cholestoral',
             c = 'target',
             ax=subfig)
plt.show()
```

# Entire life cycle of Object Oriented Plot

```python
#create figure and graph are
fig , ax = plt.subplots(figsize = (10,6))
#plot data
scatter_plot = ax.scatter(x = above_55['age'], 
                         y = above_55['cholestoral'],
                         c = above_55['target'])      # legend assigning
#coustomizations
ax.set(title='Suspectability of Heart Disease',
      xlabel='Age',
      ylabel= 'Cholestrol')
#set legend
ax.legend(*scatter_plot.legend_elements(), title = 'target')
plt.show()
```

# Entire life cycle of Object Oriented Plot Advanced

```python
#create plot area 
fig,(plot1,plot2)=plt.subplots(figsize= (10,10), 
                             nrows=2, ncols=1)
#add data
scatter_plot = plot1.scatter(x=above_55['age'], 
                            y = above_55['cholestoral'], 
                            c = above_55['target'])
#costomization
plot1.set(title='Heart Disease Problem' ,  
         xlabel='Age', 
         ylabel='Cholestrol')
#set legend
plot1.legend(*scatter_plot.legend_elements(),title= 'target')

################## Now lets plot second pot ########################
scatter_plot = plot2.scatter(x=above_55['age'], 
                            y= above_55['resting_blood_pressure'] , 
                            c = above_55['target'])
plot2.set(title='Heart Disease Problem', 
         xlabel = 'Age', 
         ylabel= 'Resting Blood Pressure')
plot2.legend(*scatter_plot.legend_elements(),title = 'target')
plt.show()
```

```python
#1-create plot area 
fig,(plot1,plot2)=plt.subplots(figsize= (10,10), 
                             nrows=2, ncols=1,
                               sharex=True) #sharex shares x-axis
#2a-add data
scatter_plot = plot1.scatter(x=above_55['age'], 
                            y = above_55['cholestoral'], 
                            c = above_55['target'], 
                            cmap= 'winter')  #cmap color the plots
#3a- coustomizations 
plot1.set(title='Heart Disease Problem' ,  
         # xlabel='Age', # it is commented because x-axis is shared
         ylabel='Cholestrol')
#4a-set legend
plot1.legend(*scatter_plot.legend_elements(),title= 'target')
#5a-meanline
plot1.axhline(y=above_55['cholestoral'].mean(),linestyle='--')

################## Now lets plot second pot ########################
#2b- add data
scatter_plot = plot2.scatter(x=above_55['age'], 
                            y= above_55['resting_blood_pressure'] , 
                            c = above_55['target'], 
                            cmap= 'plasma')  #cmap color the plots
#3b- coustomization
plot2.set(title='Heart Disease Problem', 
         xlabel = 'Age', 
         ylabel= 'Resting Blood Pressure')
#4b- set legend
plot2.legend(*scatter_plot.legend_elements(),title = 'target')
#5b-meanline
plot2.axhline(y=above_55['resting_blood_pressure'].mean(),linestyle='--')
#6-color data 
#7-color plots
plt.style.use('seaborn-v0_8-darkgrid')
#8-Naming figure
fig.suptitle('Parients Data', fontsize = 32 ,fontweight = 'bold' )
#9-show function
plt.show()
#saving the figure
fig.savefig('patient data report.png')
```
